```python
# from mylab import *
from deepar.dataset.time_series import MockTs
from deepar.dataset.time_series import TimeSeries
from deepar.dataset.time_series import NasaTs
from deepar.model.lstm import DeepAR
import math

%matplotlib inline
from numpy.random import normal
import tqdm
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

# read series 

```python
# read series
SIMULATE = False
OFFSET = 1.8
# OFFSET = 9  # for max

if SIMULATE:
    # simulate data
    l = []
    for i in range(400):
        y = 1.005**i+np.random.standard_normal()*0.1
        l.append(y) 
    s = pd.Series(l)
    
else:
    train_df = pd.read_csv("./temp/result_filtered.csv")
    s = train_df.ts_filtered
    s = s - OFFSET

print(s.shape)
plt.figure(figsize = (16,5))
plt.plot(s)
```

# train

```python
# load data and define model
# ts = MockTs()
EPOCH = 50
TIME_STEPS = 50
BATCH_SIZE = 10
STEPS_PER_BATCH = math.ceil(len(s-TIME_STEPS)/(BATCH_SIZE))

# construct dataset and train
ts = NasaTs(s, n_steps = TIME_STEPS)
dp_model = DeepAR(ts, epochs=EPOCH, batch_size = BATCH_SIZE)
dp_model.instantiate_and_fit()
```

# predict

```python
# predict all data
n_steps = TIME_STEPS
all_df = pd.DataFrame()
for i in range(len(s)-(n_steps+1)):
    batch_index = np.arange(i,i+1)
    batch = ts.get_one_batch(batch_index, n_steps)  # next_batch(self, batch_num, n_steps):
    tot_res = get_predict(dp_model,batch)
    all_df = pd.concat([all_df,tot_res], axis = 0)
```

```python
# plot and save 
predict_df = all_df.loc[all_df.index.values == n_steps-1, :]
print(predict_df.shape)
predict_df = predict_df.reset_index(drop = True)
predict_df.to_csv("./output/result_train_DeepAR.csv",index = False)
predict_df.head()
plt.figure(figsize = (16,6))
plot_predict(predict_df)
plt.plot(np.arange(len(s[TIME_STEPS:])), s[TIME_STEPS:])
```
