```python
# read predict data
train_df = pd.read_csv("temp/train_most_effective_cols.csv")
s = train_df.down_collectorEmitterVoltage_llr
# simulate data
l = []
for i in range(400):
    y = 1.005**i+np.random.standard_normal()*0.1
    l.append(y) 
s = pd.Series(l)
```

```python
# load data and define model
# ts = MockTs()
EPOCH = 100
TIME_STEPS = 50
BATCH_SIZE = 10
STEPS_PER_BATCH = math.ceil(len(s-TIME_STEPS)/(BATCH_SIZE))

# construct dataset and train
ts = NasaTs(s, n_steps = TIME_STEPS)
dp_model = DeepAR(ts, epochs=EPOCH, batch_size = BATCH_SIZE)
dp_model.instantiate_and_fit()
```

# predict 

```python
from deepar.mypredict import get_predict
from deepar.myplot import plot_batch
from deepar.myplot import plot_predict
```

```python
# try one batch
batch_num = 200
batch_index = np.arange(batch_num,batch_num+1)
batch = ts.get_one_batch(batch_index, TIME_STEPS)  # next_batch(self, batch_num, n_steps):
tot_res = get_predict(dp_model,batch)
plot_batch(batch,TIME_STEPS)
plot_predict(tot_res)
```

```python
# predict all data
n_steps = TIME_STEPS
all_df = pd.DataFrame()
for i in range(len(s)-(n_steps+1)):
    batch_index = np.arange(i,i+1)
    batch = ts.get_one_batch(batch_index, n_steps)  # next_batch(self, batch_num, n_steps):
    tot_res = get_predict(dp_model,batch)
    all_df = pd.concat([all_df,tot_res], axis = 0)
```

```python
# plot and save 
predict_df = all_df.loc[all_df.index.values == n_steps-1, :]
print(predict_df.shape)
predict_df = predict_df.reset_index(drop = True)
predict_df.to_csv("./output/predict_DeepAR.csv",index = False)
predict_df.head()
plt.figure(figsize = (16,6))
plot_predict(predict_df)
plt.plot(np.arange(len(s[TIME_STEPS:])), s[TIME_STEPS:])
```
